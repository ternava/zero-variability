### Replicate-Experiments Notebook

This notebook automates the replication of the finding in our paper, as follows:

1. Clones required repositories (Linux kernel, Coreutils, Tree-sitter C grammar)  
2. Builds and loads the Tree-sitter C parser  
3. Reads measurement data (`final2.csv`) for the 108 Coreutils programs  
4. Scans all Bash scripts under `linux/` to count how often each program name appears  
5. Categorizes usage levels (low/medium/high) based on word counts  
6. Parses Coreutils source files to extract compile-time option macros  
7. Exports the enriched data (`final4.csv`)  
8. Generates bubble plots and evolution charts to visualize size, options, compile-time flags, and LOC growth  
9. Computes Pearson and Spearman correlations between metrics  

Run cells in order, ensuring that the `linux`, `coreutils`, and `tree-sitter-c` directories are present and built.

```python
# Import all necessary libraries
# make sure to install the required packages, such as (see the file requirements.txt too): 
# pip install numpy pandas matplotlib scipy tree-sitter

# Standard library
from pathlib import Path

# Third-party
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from matplotlib.patches import Rectangle, FancyArrowPatch
from matplotlib.ticker import ScalarFormatter
from scipy.stats import spearmanr

# Tree-sitter
from tree_sitter import Language, Parser
```

```python
# Builds and loads the Tree-sitter C parser
import ctypes

# Load your compiled .so
lib = ctypes.CDLL('tree-sitter-c/build/my-languages.so')

# Grab the exported `tree_sitter_c` symbol and tell ctypes it's returning a void*
c_fn = lib.tree_sitter_c
c_fn.restype = ctypes.c_void_p

# Call it and wrap in c_void_p to avoid the deprecation warning
raw_ptr = c_fn()
ptr = ctypes.c_void_p(raw_ptr)
LANGUAGE = Language(raw_ptr) #(ptr.value)

# Init the parser
parser = Parser()
parser.language = LANGUAGE

print("Tree-sitter C language loaded successfully!")
```

```python
# Reads measurement data (`final2.csv`) for the 108 Coreutils programs
df = pd.read_csv("../measures/final2.csv")
df = df.dropna()
df.head()
```

```python
# Scans all Bash scripts under `linux/` to count how often each program name appears
# it should be noted that the cloned linux may provide you different results

linux_dir = Path('linux').expanduser()
all_bash_files = list(linux_dir.rglob('*.sh'))
print("Found", len(all_bash_files), "bash scripts")
```

```python
# Count the total words in all bash scripts
words = [
    w
    for path in all_bash_files
    for w in Path(path).read_text(errors='ignore').split()
]
print("Total words:", len(words))
```

```python
# Categorizes usage levels (low/medium/high) based on word counts
# 2 = high usage, ofetn used
# 1 = medium, not so often but still
# 0 = few utilisations of the coreutils function

from collections import Counter

word_counts = Counter(words)
df['countw'] = df['program'].map(word_counts)
df['usage'] = pd.cut(
    df['countw'],
    bins=[-1, 50, 1000, np.inf],
    labels=[0, 1, 2]
).astype(int)
```

```python
# Parses Coreutils source files to extract compile-time option macros 
#Augments the DataFrame with:
#   - `countw`: script-usage count  
#   - `usage`: binned usage category  
#   - `comp`: number of compile-time options 

def extract_macros(node, src_bytes):
    comps = set()
    for c in node.children:
        if c.type in ("binary_expression","unary_expression","identifier"):
            txt = src_bytes[c.start_byte:c.end_byte].decode()
            comps.add(txt.replace('!','').replace('defined','').strip())
    return comps

core = Path('coreutils/src')
opt_comp = {prog: set() for prog in dep_graph}

for prog, files in dep_graph.items():
    for fname in files:
        src = (core/fname).read_bytes() if (core/fname).exists() else None
        if not src:
            print("Missing:", fname)
            continue
        root = parser.parse(src).root_node
        for child in root.children:
            if child.type.startswith('preproc_'):
                opt_comp[prog] |= extract_macros(child, src)

df['comp'] = df['program'].map(lambda p: len(opt_comp.get(p,())))
```

```python
dep_graph = {'[' : ['test.c'],
'arch' : ['uname-arch.c', 'uname.c'],
'b2sum' : ['digest.c'],
'base32' : ['basenc.c'],
'base64' : ['basenc.c'],
'basename' : ['basename.c'],
'basenc' : ['basenc.c'],
'cat' : ['cat.c'],
'chcon' : ['chcon.c'],
'chgrp' : ['chgrp.c'],
'chmod' : ['chmod.c'],
'chown' : ['chown.c'],
'chroot' : ['chroot.c'],
'cksum' : ['cksum.c','digest.c'],
'comm' : ['comm.c'],
'cp' : ['cp.c'],
'csplit' : ['csplit.c'],
'cut' : ['cut.c'],
'date' : ['date.c'],
'dd' : ['dd.c'],
'df' : ['df.c'],
'dir' : ['ls-dir.c', 'ls.c'],
'dircolors' : [ 'dircolors.c'],
'dirname' : [ 'dirname.c'],
'du' : [ 'du.c'],
'echo' : [ 'echo.c'],
'env' : [ 'env.c'],
'expand' : [ 'expand.c'],
'expr' : [ 'expr.c'],
'factor' : [ 'factor.c'],
'false' : ['false.c', 'true.c'],
'fmt' : [ 'fmt.c'],
'fold' : [ 'fold.c'],
'groups' : [ 'groups.c'],
'head' : [ 'head.c'],
'hostid' : [ 'hostid.c'],
'hostname' : [ 'hostname.c'],
'id' : [ 'id.c'],
'install' : [ 'install.c'],
'join' : [ 'join.c'],
'kill' : [ 'kill.c'],
'link' : ['force-link.c', 'link.c'],
'ln' : [ 'ln.c'],
'logname' : [ 'logname.c'],
'ls' : ['ls-ls.c', 'ls.c'],
'md5sum' : [ 'digest.c'],
'mkdir' : [ 'mkdir.c'],
'mkfifo' : [ 'mkfifo.c'],
'mknod' : [ 'mknod.c'],
'mktemp' : [ 'mktemp.c'],
'mv' : [ 'mv.c'],
'nice' : [ 'nice.c'],
'nl' : [ 'nl.c'],
'nohup' : [ 'nohup.c'],
'nproc' : [ 'nproc.c'],
'numfmt' : [ 'numfmt.c'],
'od' : [ 'od.c'],
'paste' : [ 'paste.c'],
'pathchk' : [ 'pathchk.c'],
'pinky' : [ 'pinky.c'],
'pr' : [ 'pr.c'],
'printenv' : [ 'printenv.c'],
'printf' : [ 'printf.c'],
'ptx' : [ 'ptx.c'],
'pwd' : [ 'pwd.c'],
'readlink' : [ 'readlink.c'],
'realpath' : [ 'realpath.c'],
'rm' : [ 'rm.c'],
'rmdir' : [ 'rmdir.c'],
'runcon' : [ 'runcon.c'],
'seq' : [ 'seq.c'],
'sha1sum' : [ 'digest.c'],
'sha224sum' : [ 'digest.c'],
'sha256sum' : [ 'digest.c'],
'sha384sum' : [ 'digest.c'],
'sha512sum' : [ 'digest.c'],
'shred' : [ 'shred.c'],
'shuf' : [ 'shuf.c'],
'sleep' : [ 'sleep.c'],
'sort' : [ 'sort.c'],
'split' : [ 'split.c'],
'stat' : [ 'stat.c'],
'stdbuf' : [ 'stdbuf.c'],
'stty' : [ 'stty.c'],
'sum' : ['sum.c', 'digest.c'],
'sync' : [ 'sync.c'],
'tac' : [ 'tac.c'],
'tail' : [ 'tail.c'],
'tee' : [ 'tee.c'],
'test' : [ 'test.c'],
'timeout' : [ 'timeout.c'],
'touch' : [ 'touch.c'],
'tr' : [ 'tr.c'],
'true' : [ 'true.c'],
'truncate' : [ 'truncate.c'],
'tsort' : [ 'tsort.c'],
'tty' : [ 'tty.c'],
'uname' : ['uname-uname.c','uname.c'],
'unexpand' : [ 'unexpand.c'],
'uniq' : [ 'uniq.c'],
'unlink' : [ 'unlink.c'],
'uptime' : [ 'uptime.c'],
'users' : [ 'users.c'],
'vdir' : ['ls-vdir.c','ls.c'],
'wc' : [ 'wc.c'],
'who' : [ 'who.c'],
'whoami' : [ 'whoami.c'],
'yes' : [ 'yes.c']}


# we cloned coreutils in the parent directory 
# git clone https://git.savannah.gnu.org/git/coreutils.git
all_c_files =[]
for dep in dep_graph:
    all_c_files.extend(dep_graph[dep])

core_dir='coreutils/src/'

# sanity check to test if all files are present
missing_files = [ar for ar in all_c_files if not os.path.exists(os.path.join(core_dir, ar))]
if missing_files:
    print(f"Warning: Missing files: {missing_files}. Skipping missing files.")

def get_code(source, node):
    ### input : a node of the ast
    ### output : get the "code" of the node of the ast, 
    ### i.e. the string content of the related part in the code

    code = source[node.start_byte:node.end_byte].decode('utf8')
    code = code.replace('\n', '').replace('\t','')
    return code
    
nb_comp = dict()
opt_comp = dict()

for dg in dep_graph:
    nb_comp[dg] = 0
    opt_comp[dg] = []

for dg in dep_graph:
    for c_path in dep_graph[dg]:
        
        comps = []

        s = ""
        full_path = os.path.join(core_dir, c_path)
        if not os.path.exists(full_path):
            print(f"Skipping missing file: {full_path}")
            continue
        with open(full_path, "r") as f:
            s += f.read() + "\n"

        #print('#if' in s)
        #print('#ifndef' in s)

        source = bytes(s, "utf8")
        ast = parser.parse(source)
        root_node = ast.root_node

        for child in root_node.children:
            if child.type in ["preproc_if", "preproc_def", "preproc_ifdef"]:
                for n in child.children:
                    if n.type == 'binary_expression':
                        for nc in n.children:
                            if nc.type in ['unary_expression', 'identifier']:
                                new_comp = get_code(source, nc).replace('!',
                                                                       '').replace('defined', 
                                                                                   '').replace(' ', 
                                                                                               '')
                                if new_comp not in comps:
                                    comps.append(new_comp)
                    if n.type in ['unary_expression', 'identifier']:
                        new_comp = get_code(source, n).replace('!','').replace('defined', '').replace(' ', '')
                        if new_comp not in comps:
                            comps.append(new_comp)
        opt_comp[dg].extend(comps)

for oc in opt_comp:
    nb_comp[oc] = len(opt_comp[oc])
    
df['comp'] = [nb_comp[oc] for oc in df['program']]
```
